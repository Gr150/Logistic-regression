```python
#Data preparation

#There are 3 columns named default, housing, loan, y having only 2 values yes or no. 
#We are converting it to 0 and 1

# Select the categorical columns
columns_yn = ["default", "housing", "loan", "y"]

# Map the "yes" and "no" values to 1 and 0, respectively
df[columns_yn] = df[columns_yn].replace({"yes": 1, "no": 0})
```

```python
df
```

```python
# Select the columns and check their unique values
cols = ["job", "marital", "education","contact","month","poutcome"]

# Apply the value_counts method to each column
value_counts = df[cols].apply(lambda x: x.unique())

# Print the value counts for each column
print(value_counts)
```

```python
df["job"].unique()
```

```python
df["month"].unique()
```

```python
# Select the categorical columns marital and perform encoding
#columns_marital = ["married", "single", "divorced"]

# Map the "married","single" and "divorced" values to 0,1 and  respectively
df["marital"] = df["marital"].replace({"married": 0, "single": 1, "divorced":2})
```

```python
df
```

```python
# Select the categorical columns education and perform encoding


# Map the "unknown","primary","secondary" and "tertiary" values to 0,1,2 and 3 respectively
df["education"] = df["education"].replace({"unknown": 0, "primary": 1, "secondary":2, "tertiary":3})
```

```python
# Select the categorical columns contact and perform encoding


# Map the "unknown","cellular","telephone" values to 0,1,2 spectively
df["contact"] = df["contact"].replace({"unknown": 0, "cellular": 1, "telephone":2})
```

```python
# Select the categorical columns poutcome and perform encoding


# Map the "unknown","success","failure" and "other" values to 0,1,2 and 3 respectively
df["poutcome"] = df["poutcome"].replace({"unknown": 0, "success": 1, "failure":2,"other":3})
```

```python
#perform one hot encoding for the column job and month

# Select the columns to encode
cols_to_encode = ["job", "month"]

# Perform one-hot encoding on the columns
df = pd.get_dummies(df, columns=cols_to_encode)
```

```python
#model building

#seperate input(X) and output(Y)

X=df.drop('y',axis=1)

Y=df[['y']]
```

```python
X
```

```python
#perform model validation techniques
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.3,random_state=123,shuffle=True)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model= LogisticRegression()
```

```python
logistic_model.fit(X_train,Y_train)
```

```python
#model testing and evaluation
```

```python
#training accuracy
y_predicted_train=logistic_model.predict(X_train)
```

```python
#testing accuracy
y_predicted_test=logistic_model.predict(X_test)
```

```python
#model validation

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_train, y_predicted_train)

```

```python
accuracy_score(Y_test, y_predicted_test)
```
